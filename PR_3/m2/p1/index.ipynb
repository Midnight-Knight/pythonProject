```python
import numpy as np
import matplotlib.pyplot as plt
pairs = '. . L E X E G E Z A C E B I S O U S E S A R M A I N D I R E A . E R A T E N B E R A L A V E T I E D O R Q U A N T E I S R I O N'.split(' ')
class plansys:
    def __init__(self):
        self.x = 0
        self.y = 0
        self.economy = 0
        self.govtype = 0
        self.techlev = 0
        self.population = 0
        self.productivity = 0
        self.radius = 0
        self.goatsoupseed = [0,0,0,0]
        self.name = ['.']*12


def tweakseed(s):
    temp = (s[0] + s[1] + s[2]) & 0xFFFF # 2 byte arithmetic
    s[0], s[1], s[2] = s[1], s[2], temp
    return s

def makesystem(s):
    thissys = plansys()
    pair1, pair2, pair3, pair4 = 0, 0, 0, 0
    longnameflag = s[0] & 64
    thissys.x = s[1] >> 8
    thissys.y = s[0] >> 8

    thissys.govtype = (s[1] >> 3) & 7
    thissys.economy = (s[0] >> 8) & 7
    if thissys.govtype <= 1:
        thissys.economy = thissys.economy | 2

    thissys.techlev = ((s[1] >> 8) & 3) + (thissys.economy ^ 7)
    thissys.techlev += thissys.govtype >> 1
    if (thissys.govtype & 1) == 1:
        thissys.techlev += 1

    thissys.population = 4 * (thissys.techlev) + (thissys.economy)
    thissys.population += (thissys.govtype) + 1

    thissys.productivity = (((thissys.economy) ^ 7) + 3)*((thissys.govtype) + 4)
    thissys.productivity *= (thissys.population) * 8

    thissys.radius = 256 * (((s[2] >> 8) & 15) + 11) + thissys.x

    thissys.goatsoupseed[0] = s[1] & 0xFF
    thissys.goatsoupseed[1] = s[1] >> 8
    thissys.goatsoupseed[2] = s[2] & 0xFF
    thissys.goatsoupseed[3] = s[2] >> 8

    pair1 = 2 * ((s[2] >> 8) & 31)
    tweakseed(s)
    pair2 = 2 * ((s[2] >> 8) & 31)
    tweakseed(s)
    pair3 = 2 * ((s[2] >> 8) & 31)
    tweakseed(s)
    pair4 = 2 * ((s[2] >> 8) & 31)
    tweakseed(s)


    (thissys.name)[0] = pairs[pair1]
    (thissys.name)[1] = pairs[pair1 + 1]
    (thissys.name)[2] = pairs[pair2]
    (thissys.name)[3] = pairs[pair2 + 1]
    (thissys.name)[4] = pairs[pair3]
    (thissys.name)[5] = pairs[pair3 + 1]

    if (longnameflag):
        (thissys.name)[6] = pairs[pair4]
        (thissys.name)[7] = pairs[pair4 + 1]
        (thissys.name)[8] = '.'
    thissys.name = [i for i in thissys.name if i != '.']
    return thissys,s;

arr = [0x5A4A, 0x0248, 0xB753]
fig = plt.figure()
fig.set_figheight(15)
fig.set_figwidth(30)
ax = fig.add_subplot()
ax.axis([-10,266,266,-10])
for i in range(0,256):
    sys, arr = makesystem(arr)
    print(sys.__dict__)
    ax.plot(sys.x,sys.y,'ko')
    ax.annotate(sys.name[0],xy=(sys.x,sys.y),xytext=(sys.x+2,sys.y-2))
```

```python
import random
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

# загрузка изображения
img = Image.open('./../../image.jpg')
img_array = np.array(img)

# генерация случайных точек
num_points = 400
points = [(random.randint(0, img_array.shape[1]), random.randint(0, img_array.shape[0])) for i in range(num_points)]

# вычисление ближайших точек и закрашивание пикселей цветом этих точек
voronoi = np.zeros_like(img_array)
for x in range(img_array.shape[1]):
    for y in range(img_array.shape[0]):
        dists = [((x - p[0]) ** 2 + (y - p[1]) ** 2, i) for i, p in enumerate(points)]
        min_dist, min_idx = min(dists)
        voronoi[y, x] = img_array[points[min_idx][1], points[min_idx][0]]

# построение диаграммы Вороного
fig, ax = plt.subplots(figsize=(12, 12))
ax.imshow(voronoi)
for p in points:
    ax.plot(p[0], p[1], 'ro')
plt.show()
```
